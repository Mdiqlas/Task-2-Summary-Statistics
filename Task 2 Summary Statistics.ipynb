### **Dropping the row when whole row is null**

```python
df.dropna(how="all")
```

## **Dropping the null values in the column 'Embarked'**

```python
df.dropna(subset=['Embarked'], inplace=True)
```

### **Counting the No of unique values in the Cabin column of the Dataframe**

```python
df["Cabin"].value_counts()
```

### **Calculating the mode of Cabin column**

```python
df["Cabin"].mode()
```

### **Replacing the missing values with mode values in Cabin column**

```python
df['Cabin'].fillna(df['Cabin'].mode()[2],inplace=True)
```

#**Removing the missing data from age column**

Calculates the first quartile (Q1) and the third quartile (Q3) of the 'Age' column in a DataFrame.Calculates the interquartile range (IQR) for a dataset which representing the range between the first quartile (Q1) and the third quartile (Q3).

```python
Q1=df.Age.quantile(0.25)
Q3=df.Age.quantile(0.75)

Q1,Q3

IQR=Q3-Q1
IQR

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

### **Identifing the data without oulliers**


```python
no_outlier=df[(df.Age>=lower_limit)&(df.Age<=upper_limit)]
no_outlier
```

## **Calculating the mean of Age column**

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```
